                                      #Changing Data Types of Columns from object to float

```python
#changing type of discount price, actual price, discounted percentage and rating

amazon_df['discounted_price'] = amazon_df['discounted_price'].str.replace("₹",'')
amazon_df['discounted_price'] = amazon_df['discounted_price'].str.replace(",",'')
amazon_df['discounted_price'] = amazon_df['discounted_price'].astype('float64')


amazon_df['actual_price'] = amazon_df['actual_price'].str.replace("₹",'')
amazon_df['actual_price'] = amazon_df['actual_price'].str.replace(",",'')
amazon_df['actual_price'] = amazon_df['actual_price'].astype('float64')


amazon_df['discount_percentage'] = amazon_df['discount_percentage'].str.replace("%",'').astype('float64')
amazon_df['discount_percentage'] = amazon_df['discount_percentage']/100


```

```python
#Unusual in column rating
amazon_df['rating'].value_counts()
```

```python
# Changing Rating and rating count Columns Data Type and replacing unusual
amazon_df['rating'] = amazon_df['rating'].str.replace("|",'3.9').astype('float64')
amazon_df['rating_count'] = amazon_df['rating_count'].str.replace(",",'').astype('float64')

```

# Heatmap

Changing encoding for heatmap to numerical

```python
from sklearn.preprocessing import LabelEncoder

le_product_id = LabelEncoder()
le_category = LabelEncoder()
le_review_id = LabelEncoder()
le_review_content = LabelEncoder()
le_product_name = LabelEncoder()
le_user_name = LabelEncoder()
le_about_product = LabelEncoder()
le_user_id = LabelEncoder()
le_review_title = LabelEncoder()
le_img_link = LabelEncoder()
le_product_link = LabelEncoder()


amazon_df['product_id'] = le_product_id.fit_transform(amazon_df['product_id'])
amazon_df['category'] = le_category.fit_transform(amazon_df['category'])
amazon_df['review_id'] = le_review_id.fit_transform(amazon_df['review_id'])
amazon_df['review_content'] = le_review_content.fit_transform(amazon_df['review_content'])
amazon_df['product_name'] = le_product_name.fit_transform(amazon_df['product_name'])
amazon_df['user_name'] = le_user_name.fit_transform(amazon_df['user_name'])
amazon_df['about_product'] = le_about_product.fit_transform(amazon_df['about_product'])
amazon_df['user_id'] = le_user_id.fit_transform(amazon_df['user_id'])
amazon_df['review_title'] = le_review_title.fit_transform(amazon_df['review_title'])
amazon_df['img_link'] = le_img_link.fit_transform(amazon_df['img_link'])
amazon_df['product_link'] = le_product_link.fit_transform(amazon_df['product_link'])
```

```python
plt.figure(figsize = (16,10))
corr = amazon_df.corr()
sns.heatmap(corr, annot = True, cmap = 'coolwarm')
plt.tight_layout()
```

## What is the distribution of discounted prices vs. actual prices?

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize = (10,5))

ax1.hist(amazon_df['discounted_price'])
ax1.set_title('Distribution of Discounted Prices')
ax1.set_xlabel('Price')
ax1.set_ylabel('Product')

ax2.hist(amazon_df['actual_price'])
ax2.set_title('Distribution of Actual prices')
ax2.set_xlabel('Price')
ax2.set_ylabel('Product')

amazon_df["discount_percentage"] = (amazon_df["actual_price"] - amazon_df["discounted_price"]) / amazon_df["actual_price"] * 100
ax3.hist(amazon_df['discount_percentage'])
ax3.set_title('Distribution of Discount Percentage')
ax3.set_xlabel('Price')
ax3.set_ylabel('Product')

plt.suptitle('Distribution of Discounted prices vs Actual prices', fontsize = 14, y = 1.03, fontweight = 'bold')

plt.tight_layout()


```

*  The output shows that discounted prices are generally lower than actual prices, with a median discounted price of Rs.200 and a median actual price of Rs.400.

* The discount percentage distribution is skewed to the left, with most products having a discount of 30% or less.

* The output suggests that there may be opportunities to increase discounted prices or discount percentages to attract more customers.

```python
avg_discount_per_category = amazon_df.groupby('category')['discount_percentage'].mean().reset_index()
print(f"The average discount percentage accross category : \n{avg_discount_per_category}")
```

* Categories 1 and 3 stand out with notably higher average discounts (78.39% and 56.34%), suggesting potential factors like clearance efforts, high competition, or lower-profit margins.
* Categories 0, 206, 207, 210 have average discounts of 0%, indicating consistent pricing or strong demand for products within those categories.
* Other categories exhibit varying discount percentages, likely reflecting diverse pricing strategies and market dynamics.

### Q6: What are the most popular product keywords?

```python
amazon_df.columns
```

```python
import pandas as pd
import re
from collections import Counter

#Combining two columns product name and about product and converting type

text_data = amazon_df['product_name'].astype(str) + " " + amazon_df['about_product'].astype(str)
text_data = text_data.str.lower()

# Remove punctuation and split into words
words = re.findall(r'\b[a-zA-Z]{3,}\b', ' '.join(text_data))

#count_frequencies
word_counts = Counter(words)


# Get top 20
top_keywords = word_counts.most_common(20)

# Convert to DataFrame
top_keywords_df = pd.DataFrame(top_keywords, columns=['Keyword', 'Frequency'])

# Display the DataFrame
print(f"Most populare keyword are :\n{top_keywords_df}")


```

```python
#Optional Word Cloud

from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(words))
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Popular Product Keywords')
plt.show()
```

* USB connectivity, charging (especially fast charging), and cables are prominent product features.
* Prepositions and conjunctions like "with", "for", "and", "to" suggest a focus on explaining product compatibility and usage scenarios.
* Cables and smart devices are likely well-represented in the dataset.
* Product names tend to be concise and use common words, potentially benefiting from refined keyword extraction techniques.

## Q7: What are the most popular product reviews?

```python
from textblob import TextBlob
amazon_df[['product_id', 'user_id','review_content']]

#Calculate sentiment score
amazon_df['sentiment'] = amazon_df['review_content'].apply(lambda text: TextBlob(text).sentiment.polarity)

#sort by sentiment score
amazon_sorted =  amazon_df.sort_values(by = 'sentiment',ascending = True)

#selecting top reviews
top_reviews = amazon_sorted.head(10)

print(f"Most popular procut review are : \n{top_reviews}")
```

* The overall sentiment scores are relatively low, suggesting a tendency towards neutral or slightly negative reviews in the sample.

* The review with the highest sentiment score is "I have installed this in my kitchen working fine" (product_id 1463) with a score of -0.170167, indicating a mildly positive sentiment.

* The review with the lowest sentiment score is "tv on off not working, so difficult to battery charge" (product_id 155) with a score of -0.600000, suggesting a strongly negative sentiment.

* Several reviews mention issues with battery charging (product_id 155), product quality (product_id 1237), and ease of use (product_id 1198), highlighting potential areas for improvement.

* Some reviews express both positive and negative aspects within the same text, like "Like and happy,,Please don't buy this heater" (product_id 1237), suggesting a nuanced evaluation of the product.

* The user_id column seems to contain commas, indicating multiple user IDs for some reviews. This might need investigation to ensure accuracy.

* Reviews for product_id 22, 152, and 723 have identical content, suggesting potential data duplication or errors.

### Q8: What is the correlation between discounted_price and rating?

```python
corr = amazon_df['discounted_price'].corr(amazon_df['rating'])
print(f"The correlation between discounted price and rating is : {corr:.2f}")
```

Discounted price and rating have a weak positive correlation. This means that products with higher discounted prices tend to have slightly higher ratings, but the relationship is not very strong.

### Q9: What are the Top 5 categories based with highest ratings?

```python
category_rating = amazon_df.groupby('category')['rating'].mean()
top_5_categories = category_rating.sort_values(ascending = False).head(5)

print(f"Top 5 categories based on highest rating are \n{top_5_categories}")
```

* The top 5 categories have average ratings between 4.50 and 4.60, indicating overall positive customer satisfaction within these areas.
* Most of the top-rated categories fall within technology-related domains, including tablets, networking devices, photography accessories, media streaming devices, and calculators.
* Within broader categories like "Computers & Accessories" and "Electronics," specific subcategories emerge as particularly well-rated, such as tablets, powerline adapters, film accessories, and streaming clients.
* Four categories share a rating of 4.50, suggesting similar levels of customer satisfaction across these areas.
* The presence of "Memory" in the top 5 suggests that even relatively simple products can achieve high ratings if they meet customer needs effectively.
